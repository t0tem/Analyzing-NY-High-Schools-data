```python
import pandas as pd
```

```python
data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]
```

```python
data = {}

for n in data_files:
    key = n.replace(".csv", "")
    data[key] = pd.read_csv("schools/" + n)
```

```python
all_survey = pd.read_csv("schools/survey_all.txt", encoding="windows-1252", delimiter="\t")
d75_survey = pd.read_csv("schools/survey_d75.txt", encoding="windows-1252", delimiter="\t")
survey = pd.concat([all_survey, d75_survey], axis = 0)
survey.head()
```

```python
survey["DBN"] = survey['dbn']

cols = ["DBN", "rr_s", "rr_t", "rr_p", "N_s", "N_t", "N_p", 
        "saf_p_11", "com_p_11", "eng_p_11", "aca_p_11", "saf_t_11", 
        "com_t_11", "eng_t_11", "aca_t_11", "saf_s_11", "com_s_11", 
        "eng_s_11", "aca_s_11", "saf_tot_11", "com_tot_11", "eng_tot_11", "aca_tot_11"]

survey = survey.loc[:, cols]
data["survey"] = survey
data["survey"].head()

```

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]
```

```python
def pad(x):
    if len(str(x)) == 1:
        return str(x).zfill(2)
    return str(x)

data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad)
```

```python
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
```

```python
data["class_size"].head()
```

```python
cols = ['SAT Critical Reading Avg. Score',
        'SAT Math Avg. Score',
        'SAT Writing Avg. Score']

for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")

data["sat_results"]['sat_score'] = data["sat_results"][cols[0]]+data["sat_results"][cols[1]]+data["sat_results"][cols[2]]

data["sat_results"]['sat_score'].head()
```

```python
import re
def grab_lat(s):
    new_s = re.findall("\(.+\)", s)[0]
    lat = new_s.split(",")[0].replace("(","")
    return lat

data['hs_directory']['lat'] = data['hs_directory']['Location 1'].apply(grab_lat)

data['hs_directory'].head()
```

```python
def grab_lon(s):
    new_s = re.findall("\(.+\)", s)[0]
    lon = new_s.split(",")[1].replace(")","").strip()
    return lon

data['hs_directory']['lon'] = data['hs_directory']['Location 1'].apply(grab_lon)

data['hs_directory'].head()
```

```python
for c in ['lat', 'lon']:
    data['hs_directory'][c] = pd.to_numeric(data['hs_directory'][c],  errors="coerce")

data['hs_directory'].head()
```

```python
class_size = data["class_size"]

class_size = class_size[(class_size["GRADE "] == "09-12") & 
                        (class_size["PROGRAM TYPE"] == "GEN ED")]
```

```python
class_size.head()
```

we'll now aggregate df using `groupby`

```python
import numpy as np
class_size = class_size.groupby("DBN").agg(np.mean).reset_index()
data['class_size'] = class_size
```

```python
data['class_size'].head()
```

```python
demographics = data["demographics"]
demographics = demographics[demographics["schoolyear"] == 20112012]
data["demographics"] = demographics
```

```python
data["demographics"].head()
```

```python
print(
    data['graduation']['Demographic'].unique(),
    data['graduation']['Cohort'].unique(),
    sep = "\n---\n"
)
    
```

```python
graduation = data['graduation']
graduation = graduation[(graduation['Demographic'] == 'Total Cohort') &
                       (graduation['Cohort'] == '2006')]

data['graduation'] = graduation
data['graduation'].head()
```

now let's convert some columns ot numeric

```python
cols = ['AP Test Takers ',
        'Total Exams Taken',
        'Number of Exams with scores 3 4 or 5']

for c in cols:
    data['ap_2010'][c] = pd.to_numeric(data['ap_2010'][c], errors="coerce")
    print(data['ap_2010'][c].dtypes)
        
```

now we can start joining (merging) the data

```python
combined = data["sat_results"]
combined = combined.merge(data['ap_2010'], how = 'left', on = 'DBN')

combined = combined.merge(data['graduation'], how = 'left', on = 'DBN')

combined.shape
```

```python
dfs = ['class_size',
      'demographics',
      'survey',
      'hs_directory']

for df in dfs:
    combined = combined.merge(data[df], on = 'DBN', how = 'inner')

combined.shape
```

now let's fill NAs

```python
means = combined.mean()
combined = combined.fillna(means)
combined = combined.fillna(0)
combined.head()
```

```python
combined['school_dist'] = combined['DBN'].apply(lambda x: x[:2])

combined['school_dist'].head()
```
